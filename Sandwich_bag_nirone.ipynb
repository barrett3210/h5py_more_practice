# Measuring one layer of a sandwich bag over the white test blank.

The graphs in this file were created of one layer of a sandwich bag over the white test blank, using the NIRONE spectrometer.
In one set of graphs, multiple spectra were acquired with the probe stationary.
In the other set of graphs, spectra were acquired with the probe moving at a rate of 15 mm/sec.

```python
vis_moving_df = pd.DataFrame(data=vis_spectra_moving, columns=vis_waves.astype(int))

vis_stat_df = pd.DataFrame(data=vis_spectra_stationary, columns=vis_waves.astype(int))
```

## NIRONE spectrometer, stationary probe.

### Graph of all the spectra acquired.

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("Sandwich bag, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")
for i in range(vis_spectra_stationary.shape[0]):
    ax.plot(vis_waves, vis_spectra_stationary[i, :])
plt.show()
```
