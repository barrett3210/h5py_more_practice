```python
import pandas as pd
import numpy as np
import h5py
import datetime
import matplotlib.pyplot as plt

file_name = "more4/spect_test3.h5"
h5_file = h5py.File(file_name, "r")
```

# Measuring white test blank only.

The graphs in this file were created of the white test blank only, using the two Hamamatsu spectrometers.
In one set of graphs, multiple spectra were acquired with the probe stationary.
In the other set of graphs, spectra were acquired with the probe moving at a rate of 15 mm/sec

```python
ins1 = h5_file['session001/cal001/ins001']
```

```python
vis_waves = h5_file['session001/cal001'].attrs['spec1_wavelengths_vector'][:]
ftir_waves = h5_file['session001/cal001'].attrs['spec2_wavelengths_vector'][:]
```

```python
vis_spectra_stationary = h5_file[f'session001/cal001/ins001/spectrometer1/derived/absorbances'][:]
ftir_spectra_stationary = h5_file[f'session001/cal001/ins001/spectrometer2/derived/absorbances'][:]
```

```python
vis_spectra_moving = h5_file[f'session001/cal001/ins002/spectrometer1/derived/absorbances'][:]
ftir_spectra_moving = h5_file[f'session001/cal001/ins002/spectrometer2/derived/absorbances'][:]
```

```python
vis_moving_df = pd.DataFrame(data=vis_spectra_moving, columns=vis_waves.astype(int))
ftir_moving_df = pd.DataFrame(data=ftir_spectra_moving, columns=ftir_waves.astype(int))
vis_stat_df = pd.DataFrame(data=vis_spectra_stationary, columns=vis_waves.astype(int))
ftir_stat_df = pd.DataFrame(data=ftir_spectra_stationary, columns=ftir_waves.astype(int))
```

## Visible light spectrometer, stationary probe.

### Graph of all the spectra acquired.

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("White test blank, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")
for i in range(vis_spectra_stationary.shape[0]):
    ax.plot(vis_waves, vis_spectra_stationary[i, :])
plt.show()
```

### Graphs of some invidual spectra

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("White test blank, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")

ax.plot(vis_waves, vis_spectra_stationary[3, :])
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("White test blank, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")

ax.plot(vis_waves, vis_spectra_stationary[4, :])
plt.show()
```

### Graph of absorbance at 600 nm with order of acquisition

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_title("White test blank, stationary probe")
ax.set_ylabel("Acquisition number")
ax.set_xlabel("Absorbance")
plt.plot(vis_stat_df.loc[:, 600], vis_stat_df.index.values, "r-o");
```

### Graph of mean spectrum

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("Mean Spectrum, white test blank, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")

ax.plot(vis_waves, vis_spectra_stationary.mean(axis=0))
plt.show()
```

### Graph of Standard Deviation spectrum

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("Standard Deviation Spectrum, white test blank, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")

ax.plot(vis_waves, vis_spectra_stationary.std(axis=0))
plt.show()
```
