```python
import pandas as pd
import numpy as np
import h5py
import datetime
import matplotlib.pyplot as plt

file_name = "more4/nirone_test3.h5"
h5_file = h5py.File(file_name, "r")
```

# Measuring one layer of a green milk jug cap.

The graphs in this file were created of a green milk jug cap, using the NIRONE.
In one set of graphs, multiple spectra were acquired with the probe stationary.
In the other set of graphs, spectra were acquired with the probe moving at a rate of 15 mm/sec.

```python
vis_waves = h5_file['session001/cal001'].attrs['spec1_wavelengths_vector'][:]
```

```python
vis_spectra_stationary = h5_file[f'session001/cal001/ins002/spectrometer1/derived/absorbances'][:]
```

```python
vis_spectra_moving = h5_file[f'session001/cal001/ins003/spectrometer1/derived/absorbances'][:]
```

```python
vis_moving_df = pd.DataFrame(data=vis_spectra_moving, columns=vis_waves.astype(int))

vis_stat_df = pd.DataFrame(data=vis_spectra_stationary, columns=vis_waves.astype(int))
```

## NIRONE spectrometer, stationary probe.

### Graph of all the spectra acquired.

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("Green cap, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")
for i in range(vis_spectra_stationary.shape[0]):
    ax.plot(vis_waves, vis_spectra_stationary[i, :])
plt.show()
```

### Graphs of some invidual spectra

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("Green cap, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")

ax.plot(vis_waves, vis_spectra_stationary[3, :])
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title("Green cap, stationary probe")
ax.set_ylabel("Absorbance")
ax.set_xlabel("Wavelength (nm)")

ax.plot(vis_waves, vis_spectra_stationary[4, :])
plt.show()
```

### Graph of absorbance at 1600 nm with order of acquisition

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_title("Green cap, stationary probe")
ax.set_ylabel("Acquisition number")
ax.set_xlabel("Absorbance")
plt.plot(vis_stat_df.loc[:, 1600], vis_stat_df.index.values, "r-o");
```
