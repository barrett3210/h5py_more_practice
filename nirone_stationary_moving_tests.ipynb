```python
import pandas as pd
import numpy as np
import h5py
import datetime
import matplotlib.pyplot as plt

file_name = "more5/nirone_test4.h5"
h5_file = h5py.File(file_name, "r")
```

```python
waves = h5_file['session003/cal002'].attrs['spec1_wavelengths_vector'][:]
```

```python
stationary_ins = ['ins001', 'ins002', 'ins003', 'ins004', 'ins005']
moving_ins = ['ins006', 'ins007', 'ins008']
```

```python
for ins in stationary_ins:
    spectra = h5_file[f'session003/cal002/{ins}/spectrometer1/derived/absorbances'][:]
    fig, ax = plt.subplots(figsize=(10,10))
    ax.set_title(f"NIRONE, stationary, white blank, {ins}")
    ax.set_ylabel("Absorbance")
    ax.set_xlabel("Wavelength (nm)")
    for i in range(spectra.shape[0]):
        ax.plot(waves, spectra[i, :])
    plt.show()
```

```python
for ins in moving_ins:
    spectra = h5_file[f'session003/cal002/{ins}/spectrometer1/derived/absorbances'][:]
    fig, ax = plt.subplots(figsize=(10,10))
    ax.set_title(f"NIRONE, moving, white blank, {ins}")
    ax.set_ylabel("Absorbance")
    ax.set_xlabel("Wavelength (nm)")
    for i in range(spectra.shape[0]):
        ax.plot(waves, spectra[i, :])
    plt.show()
```
